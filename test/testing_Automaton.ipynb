# Iniciando con la estructura.

```
while (a < b) {
    a = b + 1;
    a / 1;
    a += 1;
    c = a * b;
}
```

De la estructura anterior se pueden rescatar algunos puntos importantes.

1. Toda estructura while comenzará con la palabra reservada _while_ en minúsculas sin importar los espacios o tabulaciones que tenga a la izquierda o derecha.
2. Toda estructura while deberá contener una operación lógica (_comparación_) entre paréntesis.
3. El cuerpo de la estructura contiene instrucciones que se repetirán hasta que el ciclo se rompa, el número de instrucciones esperadas es variable.
4. El fin de cada instrucción es denotado por un punto y coma (_;_).
5. Las operaciones pueden ser ejecutadas sin necesidad de guardarse el resultado.

````python
import PySimpleGUI as gui
import re

def while_automaton(state, chain, i,entry):

    state_changed = False

    final_states = ["P2"]
    transitions = {
        "INI" : [
            ['[w]',"W1"]
        ],
        "W1" : [
            ['[h]', "W2"]
        ],
        "W2" : [
            ['[i]', "W3"]
        ],
        "W3" : [
            ['[l]', "W4"]
        ],
        "W4" : [
            ['[e]', "P1"]
        ],
        "P1" : [
            ['[(]', "INIVA"],
        ],
        "INIVA" : [
            ['[a-zA-Z]', "VA1"]
        ],
        "VA1" : [
            ['[a-zA-Z0-9]', "VA1"],
            ['[_]', "VA2"],
            ['[=!]', "CON1"],
            ['[<>]', "CON2"]
        ],
        "VA2" : [
            ['[A-Za-z0-9]', "VA1"],
        ],
        "CON1" : [
            ['[=]', "INIVA2"]
        ],
        "CON2" : [
            ['[=]', "INIVA2"],
            ['[A-Za-z]', "VA11"]
        ],
        "INIVA2" : [
            ['[A-Za-z]', "VA11"]
        ],
        "VA11" : [
            ['[a-zA-Z0-9]', "VA11"],
            ['[_]', "VA21"],
            ['[)]', "P2"]
        ],
        "VA21" : [
            ['[A-Za-z0-9]', "VA11"]
        ]
    }

    actual_eval = transitions[state]

    for transition in actual_eval:

        if re.search(transition[0],entry)is not None:
            state = transition[1]
            state_changed = True
            break
    if not state_changed: return False
    elif i + 1 == len(chain):
        if state in final_states:
            return True
        else: return False
    i += 1
    return while_automaton(state, chain, i,chain[i])

file = open("instrucctions.txt")

content = ""

for line in file:
    content += line

container_code = gui.Multiline(content,size=(20,10))
file.close()
main_label = gui.Text("Analizador de estructuras WHILE")
btn_initializer = gui.OK()
log_container = gui.Multiline("",size=(20,10))
btn_evaluator = gui.Button("Analizar")
btn_appOut = gui.Button("Salir")

layout = [
    [main_label],
    [container_code, log_container],
    [btn_initializer,btn_evaluator,btn_appOut]
]

window_size = [400,300]

app = gui.Window("App", layout,size=window_size)

while True:

    event, values = app.read()

    if event == "OK":
        file = open("instrucctions.txt", "w")
        file.write(container_code.get())
        file.close()

    if event == "Analizar":
        chain_t_eval = container_code.get()
        if while_automaton("INI", chain_t_eval, 0,chain_t_eval[0]):
            gui.popup('Correcto')
        else:
            gui.popup("Incorrecto")
    if event == gui.WIN_CLOSED or event == "Salir":
        app.close()
        break
````
